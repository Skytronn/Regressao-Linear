```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import xlrd
from yellowbrick.regressor import ResidualsPlot
from scipy.optimize import curve_fit
import pylab
```

```python
#data = pd.read_excel('Regressao-Nao-Linear.xlsx')

x = np.array([-1.0, -0.9, -0.8, 0.0, 1.0, 2.0])
y = np.array([6.01, 5.39, 4.80, 2.01, 0.65, 0.21])
```

```python


def func(x, p1,p2):
    return p1 * np.exp(p2 * x)
```

```python
#Usando o Curve_fit
```

```python
popt, pcov = curve_fit(func, x, y,p0=(10,-1))
p1, p2 = popt
```

```python
#Parametros otimizados(ajuste)
```

```python
print('Ajuste dos parametros: p1:%g e p2:%g' % (p1,p2))
```

```python
#Plotagem do Diagrama de dispersão
```

```python
yfit = func(x, *popt)
pylab.plot(x,y,'o')
pylab.plot(x,yfit,'-')
pylab.xlabel('x')
pylab.ylabel('y')
```
