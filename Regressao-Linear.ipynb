```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import xlrd
from yellowbrick.regressor import ResidualsPlot
```

```python
#data = pd.read_csv('Psaude.csv')
data = pd.read_excel('BKP.xlsx')

X = data.iloc[:,0].values
Y = data.iloc[:,1].values
correlacao = np.corrcoef(X, Y)

X = X.reshape(-1,1)
data.corr()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X , Y)

#b0
regressor.coef_

#b1
regressor.intercept_
```

```python
plt.scatter(X, Y)
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.title("Regressão Linear Plano de Saude")
plt.plot(X, regressor.predict(X), color = 'red')

#Calculo do Score
score = regressor.score(X, Y)
print(score)
```

```python
#Visualização dos residuos

visual = ResidualsPlot(regressor)
visual.fit(X, Y)
visual.poof()
```
